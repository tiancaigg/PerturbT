# CellOracle GRN Inference: Norman 2019 (K562) with Bulk ATAC-seq

**Goal:** Build a K562-specific GRN from public bulk ATAC-seq peaks + Norman 2019 scRNA-seq → extract GRN coefficient matrix M for Stage 2 (scGPT attention bias injection).

**Pipeline:**
1. Preprocess bulk ATAC-seq BED → TSS-annotated peaks
2. TF motif scan → base GRN
3. Import Norman 2019 scRNA-seq + base GRN → Oracle object
4. GRN inference (ridge regression per cluster)
5. Export GRN coefficient matrix M

**Inputs:**
- `scratch/PerturbT/data/ATAC_Seq/K562/K562_ATAC_peaks.bed` (public K562 bulk ATAC-seq)
- `data/splits/norman_2019_raw_sparse_with_splits_original_from_GEARS.h5ad` (Norman 2019 preprocessed)

**Requirements:** `celloracle`, `scanpy`, `genomepy`, `gimmemotifs`

## 0. Setup

```python
# Install if needed (run once)
# !pip install celloracle
# celloracle installs gimmemotifs and genomepy as dependencies
```

```python
import os
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
import celloracle as co
from celloracle import motif_analysis as ma

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = [6, 4.5]
plt.rcParams['savefig.dpi'] = 300

print(f"CellOracle version: {co.__version__}")
```

```python
# === CONFIGURE PATHS ===
# Adjust these to your directory structure
BED_PATH = "../../data/ATAC_Seq/K562/K562_ATAC_peaks.bed"
NORMAN_H5AD = "../../data/splits/norman_2019_raw_sparse_with_splits_original_from_GEARS.h5ad"
OUTPUT_DIR = "./celloracle_grn/"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Reference genome for K562 (human)
REF_GENOME = "hg38"  # Change to hg19 if your BED uses hg19 coordinates
```

## 1. Bulk ATAC-seq Peak Processing → TSS Annotation

CellOracle Option 2 (bulk ATAC-seq): we get only **proximal promoter/enhancer** peaks (no Cicero distal elements, since that requires scATAC-seq).

```python
# 1.1 Load BED file
bed = ma.read_bed(BED_PATH)
print(f"Loaded {bed.shape[0]} peaks from BED file")
bed.head()
```

```python
# 1.2 Convert to peak name list
peaks = ma.process_bed_file.df_to_list_peakstr(bed)
print(f"Total peaks: {len(peaks)}")
print(f"Example: {peaks[:3]}")
```

```python
# 1.3 Install reference genome if not already installed
# Check supported genomes
print("Supported genomes:", ma.SUPPORTED_REF_GENOME.keys())
```

```python
# Install reference genome (run once, ~5-10 min)
# This downloads genome fasta + index for motif scanning
import genomepy
genomepy.install_genome(REF_GENOME, provider="UCSC")
# If already installed, this will just confirm it exists
```

```python
# 1.4 TSS annotation
# Identifies which peaks overlap with transcription start sites
# For bulk ATAC: only proximal promoter peaks are used
tss_annotated = ma.get_tss_info(peak_str_list=peaks, ref_genome=REF_GENOME)
print(f"TSS-annotated peaks: {tss_annotated.shape[0]}")
tss_annotated.head()
```

```python
# 1.5 Format for motif scan input
peak_id_tss = ma.process_bed_file.df_to_list_peakstr(tss_annotated)
tss_annotated_df = pd.DataFrame({
    "peak_id": peak_id_tss,
    "gene_short_name": tss_annotated.gene_short_name.values
}).reset_index(drop=True)

print(f"Peak-gene pairs: {tss_annotated_df.shape[0]}")
print(f"Unique genes: {tss_annotated_df['gene_short_name'].nunique()}")
tss_annotated_df.head()
```

```python
# 1.6 Save processed peaks
tss_path = os.path.join(OUTPUT_DIR, "K562_processed_peak_file.csv")
tss_annotated_df.to_csv(tss_path, index=False)
print(f"Saved to: {tss_path}")
```

## 2. TF Motif Scan → Base GRN

Scan TSS-proximal peak sequences for TF binding motifs using gimmemotifs.

**Note:** This step is computationally intensive (~30-60 min depending on peak count). The kernel may require 8-16 GB RAM.

```python
# 2.1 Load processed peak data
tss_annotated_df = pd.read_csv(os.path.join(OUTPUT_DIR, "K562_processed_peak_file.csv"))
print(f"Loaded {tss_annotated_df.shape[0]} peak-gene pairs")
```

```python
# 2.2 Instantiate TFinfo object
tfi = ma.TFinfo(
    peak_data_frame=tss_annotated_df,
    ref_genome=REF_GENOME,
    # genomes_dir="genomepy"  # uncomment if genome installed in non-default location
)
```

```python
# 2.3 Motif scan
# Uses default gimmemotifs motif database for human
# This is the rate-limiting step (~30-60 min)
%%time
tfi.scan(fpr=0.02, motifs=None, verbose=True)
# fpr=0.02 is the default false positive rate for motif scanning
```

```python
# 2.4 Save intermediate results (motif scan is expensive; save checkpoint)
tfi_path = os.path.join(OUTPUT_DIR, "K562_TFinfo.celloracle.tfinfo")
tfi.save_as_parquet(tfi_path)
print(f"Saved TFinfo to: {tfi_path}")
```

```python
# (Optional) Reload from checkpoint
# tfi = ma.load_TFinfo_from_parquet(tfi_path)
```

```python
# 2.5 Filter motif hits and reset index
tfi.reset_filtering()
tfi.filter_motifs_by_score(threshold=10)  # filter by motif score
tfi.make_TFinfo_dataframe_and_dictionary(verbose=True)
```

```python
# 2.6 Extract base GRN
# Format: dataframe with columns [peak_id, gene_short_name, TF1, TF2, ...]
# Values are binary (0/1) indicating whether TF motif is found in peak near gene
base_GRN_df = tfi.to_dataframe()
print(f"Base GRN shape: {base_GRN_df.shape}")
print(f"  Peaks x (2 index cols + {base_GRN_df.shape[1]-2} TFs)")
base_GRN_df.head()
```

```python
# 2.7 Save base GRN
base_grn_path = os.path.join(OUTPUT_DIR, "K562_base_GRN.parquet")
base_GRN_df.to_parquet(base_grn_path)
print(f"Saved base GRN to: {base_grn_path}")

# Also save as CSV for inspection
base_GRN_df.to_csv(os.path.join(OUTPUT_DIR, "K562_base_GRN.csv"), index=False)
```

## 3. Prepare Norman 2019 scRNA-seq for CellOracle

CellOracle needs:
- Raw counts in `adata.X` (or a layer)
- Clustering in `adata.obs`
- Dimensionality reduction in `adata.obsm`
- Top 2-3K highly variable genes

```python
# 3.1 Load Norman 2019 data
adata = sc.read_h5ad(NORMAN_H5AD)
print(f"Loaded: {adata.shape[0]} cells x {adata.shape[1]} genes")
print(f"obs columns: {list(adata.obs.columns)}")
adata
```

```python
# 3.2 Use only control (unperturbed) cells for GRN inference
# CellOracle infers GRNs from wild-type expression; perturbed cells would confound it
# Check what the control condition is called in your data
print(adata.obs['condition'].value_counts().head(10))
print(f"\nTotal conditions: {adata.obs['condition'].nunique()}")
```

```python
# 3.3 Subset to control cells
# Norman 2019 labels controls as 'ctrl' — verify this in your data
ctrl_mask = adata.obs['condition'] == 'ctrl'
adata_ctrl = adata[ctrl_mask].copy()
print(f"Control cells: {adata_ctrl.shape[0]}")

# If no 'ctrl' found, check for alternatives:
if adata_ctrl.shape[0] == 0:
    print("WARNING: No 'ctrl' condition found. Check condition labels:")
    print(adata.obs['condition'].unique()[:20])
```

```python
# 3.4 Ensure raw counts are in adata.X
# Check if data looks like raw counts (integers, no negatives)
import scipy.sparse as sp

if sp.issparse(adata_ctrl.X):
    sample = adata_ctrl.X[:5, :5].toarray()
else:
    sample = adata_ctrl.X[:5, :5]

print("Sample values (should be integer raw counts):")
print(sample)
print(f"\nMax value: {adata_ctrl.X.max():.1f}")
print(f"dtype: {adata_ctrl.X.dtype}")
```

```python
# 3.5 Standard scanpy preprocessing for CellOracle
# CellOracle takes raw counts but needs HVG selection + dim reduction done first

adata_co = adata_ctrl.copy()

# Store raw counts in a layer before normalizing
adata_co.layers['raw_count'] = adata_co.X.copy()

# Normalize + log1p for HVG selection and PCA
sc.pp.normalize_total(adata_co, target_sum=1e4)
sc.pp.log1p(adata_co)

# Select HVGs (CellOracle recommends 2-3K)
sc.pp.highly_variable_genes(adata_co, n_top_genes=3000)
print(f"HVGs selected: {adata_co.var['highly_variable'].sum()}")
```

```python
# 3.6 Subset to HVGs
adata_co = adata_co[:, adata_co.var['highly_variable']].copy()
print(f"After HVG filter: {adata_co.shape}")
```

```python
# 3.7 PCA + neighbors + clustering + embedding
sc.pp.scale(adata_co, max_value=10)
sc.tl.pca(adata_co, svd_solver='arpack', n_comps=50)
sc.pp.neighbors(adata_co, n_neighbors=15, n_pcs=30)

# Clustering — K562 is a single cell line, but CellOracle needs a cluster column
# Use Leiden to find sub-populations (e.g., cell cycle phases)
sc.tl.leiden(adata_co, resolution=0.5)
print(f"Leiden clusters: {adata_co.obs['leiden'].nunique()}")
print(adata_co.obs['leiden'].value_counts())
```

```python
# 3.8 Compute embedding for CellOracle visualization
sc.tl.umap(adata_co)
# Also compute draw_graph (force-directed) — CellOracle's preferred embedding
sc.tl.draw_graph(adata_co)

sc.pl.draw_graph(adata_co, color='leiden', title='K562 ctrl cells (Leiden)')
```

```python
# 3.9 Downsample if > 30K cells
if adata_co.shape[0] > 30000:
    sc.pp.subsample(adata_co, n_obs=30000, random_state=42)
    print(f"Downsampled to: {adata_co.shape[0]}")

print(f"Final shape for CellOracle: {adata_co.shape}")
```

## 4. CellOracle GRN Inference

```python
# 4.1 Instantiate Oracle object
oracle = co.Oracle()

# Import raw counts — CellOracle needs unscaled, uncentered data
# Put raw counts back into .X
adata_co.X = adata_co.layers['raw_count'].copy()

oracle.import_anndata_as_raw_count(
    adata=adata_co,
    cluster_column_name="leiden",
    embedding_name="X_draw_graph_fa"  # force-directed layout
)
print("Oracle object created.")
```

```python
# 4.2 Import base GRN (from step 2)
# If you used the ATAC-based pipeline:
base_GRN_df = pd.read_parquet(os.path.join(OUTPUT_DIR, "K562_base_GRN.parquet"))

# If you used the promoter fallback:
# base_GRN_df = co.data.load_human_promoter_base_GRN()

oracle.import_TF_data(TF_info_matrix=base_GRN_df)
print(f"Base GRN imported: {base_GRN_df.shape}")
```

```python
# 4.3 PCA for KNN imputation
oracle.perform_PCA()

# Select number of PCs
plt.plot(np.cumsum(oracle.pca.explained_variance_ratio_)[:100])
n_comps = np.where(
    np.diff(np.diff(np.cumsum(oracle.pca.explained_variance_ratio_)) > 0.002)
)[0][0]
plt.axvline(n_comps, c="k")
plt.title(f"Selected PCs: {n_comps}")
plt.xlabel("PC")
plt.ylabel("Cumulative variance explained")
plt.show()

n_comps = min(n_comps, 50)
print(f"Using {n_comps} PCs")
```

```python
# 4.4 KNN imputation
n_cell = oracle.adata.shape[0]
k = int(0.025 * n_cell)
k = max(k, 30)  # minimum k
print(f"n_cell={n_cell}, k={k}")

oracle.knn_imputation(
    n_pca_dims=n_comps,
    k=k,
    balanced=True,
    b_sight=k * 8,
    b_maxl=k * 4,
    n_jobs=4
)
print("KNN imputation done.")
```

```python
# 4.5 Save Oracle checkpoint (before expensive GRN step)
oracle_path = os.path.join(OUTPUT_DIR, "K562_norman.celloracle.oracle")
oracle.to_hdf5(oracle_path)
print(f"Checkpoint saved: {oracle_path}")
```

```python
# (Optional) Reload checkpoint
# oracle = co.load_hdf5(oracle_path)
```

```python
%%time
# 4.6 GRN inference (~30-60 min depending on cell/gene count)
# Fits ridge regression per cluster: each target gene ~ TFs in base GRN
# alpha=10 is the default regularization strength
links = oracle.get_links(
    cluster_name_for_GRN_unit="leiden",
    alpha=10,
    verbose_level=10
)
print("GRN inference complete.")
```

```python
# 4.7 Inspect GRN results
print(f"Clusters with GRNs: {list(links.links_dict.keys())}")

# Check one cluster
example_cluster = list(links.links_dict.keys())[0]
grn_df = links.links_dict[example_cluster]
print(f"\nCluster '{example_cluster}': {grn_df.shape[0]} edges")
grn_df.head(10)
```

```python
# 4.8 Filter GRN edges
links.filter_links(
    p=0.001,               # p-value threshold
    weight="coef_abs",     # rank by absolute coefficient
    threshold_number=2000  # keep top N edges per cluster
)
print("Filtered links.")
```

```python
# 4.9 Save Links object
links_path = os.path.join(OUTPUT_DIR, "K562_norman.celloracle.links")
links.to_hdf5(links_path)
print(f"Links saved: {links_path}")
```

## 5. Extract GRN Coefficient Matrix M for Stage 2

For the scGPT attention bias injection (Track A), we need a TF→target gene weight matrix.

**Strategy:** Average the cluster-specific GRN coefficients across all clusters to get a global K562 GRN, then reshape into matrix form.

```python
# 5.1 Aggregate GRN edges across clusters
all_edges = []
for cluster_name, grn in links.links_dict.items():
    grn_copy = grn.copy()
    grn_copy['cluster'] = cluster_name
    all_edges.append(grn_copy)

all_edges_df = pd.concat(all_edges, ignore_index=True)
print(f"Total edges across all clusters: {all_edges_df.shape[0]}")
all_edges_df.head()
```

```python
# 5.2 Compute mean coefficient per TF→target pair
mean_edges = all_edges_df.groupby(['source', 'target']).agg(
    coef_mean=('coef_mean', 'mean'),
    coef_abs_mean=('coef_abs', 'mean'),
    p_min=('p', 'min'),
    n_clusters=('cluster', 'nunique')
).reset_index()

print(f"Unique TF→target edges: {mean_edges.shape[0]}")
print(f"Unique TFs: {mean_edges['source'].nunique()}")
print(f"Unique targets: {mean_edges['target'].nunique()}")
mean_edges.head()
```

```python
# 5.3 Pivot to matrix form: M[gene_i, gene_j] = regulatory weight from TF_j to gene_i
# Filter for significant edges first
sig_edges = mean_edges[mean_edges['p_min'] < 0.001].copy()
print(f"Significant edges (p<0.001): {sig_edges.shape[0]}")

M_matrix = sig_edges.pivot_table(
    index='target',
    columns='source',
    values='coef_mean',
    fill_value=0.0
)

print(f"\nM matrix shape: {M_matrix.shape}")
print(f"  targets (rows) x TFs (columns)")
print(f"  Non-zero entries: {(M_matrix != 0).sum().sum()}")
print(f"  Sparsity: {1 - (M_matrix != 0).sum().sum() / M_matrix.size:.4f}")
```

```python
# 5.4 Save M matrix
M_path = os.path.join(OUTPUT_DIR, "K562_GRN_M_matrix.csv")
M_matrix.to_csv(M_path)
print(f"M matrix saved: {M_path}")

# Also save as numpy for direct loading in PyTorch
np.save(
    os.path.join(OUTPUT_DIR, "K562_GRN_M_matrix.npy"),
    M_matrix.values
)
# Save gene/TF names for index alignment
pd.Series(M_matrix.index, name='target_gene').to_csv(
    os.path.join(OUTPUT_DIR, "K562_GRN_M_target_genes.csv"), index=False
)
pd.Series(M_matrix.columns, name='source_TF').to_csv(
    os.path.join(OUTPUT_DIR, "K562_GRN_M_source_TFs.csv"), index=False
)
print("All M matrix files saved.")
```

```python
# 5.5 Also save the full edge list (useful for network analysis)
mean_edges.to_csv(
    os.path.join(OUTPUT_DIR, "K562_GRN_mean_edges.csv"),
    index=False
)
print("Edge list saved.")
```

## 6. Quick Sanity Checks

```python
# 6.1 Distribution of edge weights
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(mean_edges['coef_mean'], bins=100)
axes[0].set_title('Mean GRN coefficients')
axes[0].set_xlabel('Coefficient')

axes[1].hist(np.log10(mean_edges['p_min'] + 1e-300), bins=100)
axes[1].set_title('log10(min p-value)')
axes[1].axvline(np.log10(0.001), c='r', ls='--', label='p=0.001')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# 6.2 Top TFs by number of target genes
tf_degree = sig_edges.groupby('source')['target'].nunique().sort_values(ascending=False)
print("Top 20 TFs by out-degree (# regulated genes):")
print(tf_degree.head(20))
```

```python
# 6.3 Check overlap with Norman 2019 perturbed genes
adata_full = sc.read_h5ad(NORMAN_H5AD)
conditions = adata_full.obs['condition'].unique()
perturbed_genes = set()
for c in conditions:
    if c != 'ctrl':
        for g in c.split('+'):
            perturbed_genes.add(g.strip())

grn_tfs = set(M_matrix.columns)
overlap = perturbed_genes & grn_tfs
print(f"Norman perturbed genes: {len(perturbed_genes)}")
print(f"GRN TFs: {len(grn_tfs)}")
print(f"Overlap: {len(overlap)} ({len(overlap)/len(perturbed_genes)*100:.1f}%)")
print(f"Missing from GRN: {perturbed_genes - grn_tfs}")
```
